```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math 
from PIL import Image
```

```python
def kmeans(points, k):
    best_distance = math.inf
    best_label = []

    for attempt in range(5):
        
        initial_center_index = np.random.choice(len(points), k, replace=False)
        centers = points[initial_center_index,:]
    
        old_label = [0] * len(points)
        new_label = [0] * len(points)

        for x, point in enumerate(points):
            min_dis = math.inf
            min_index = 0
            for y, center in enumerate(centers):
                distance = 0
                for i in range(len(point)):
                    distance += (point[i] - center[i]) ** 2
                distance = distance ** 0.5
                if distance < min_dis:
                    min_dis = distance 
                    min_index = y
            old_label[x] = min_index
        
        
        count = [0] * k
        sum_r = [0] * k
        sum_g = [0] * k
        sum_b = [0] * k
        for i in range(len(old_label)):
            label = int(old_label[i])
            count[label] += 1
            sum_r[label] += points[i,0]
            sum_g[label] += points[i,1]
            sum_b[label] += points[i,2]
        for i in range(len(centers)):
            centers[i] = [sum_r[i]/count[i], sum_g[i]/count[i], sum_b[i]/count[i]]
        
        
        for times in range(0, 25):
            sum_distance = 0
            for x, point in enumerate(points):
                min_dis = math.inf
                min_index = 0
                for y, center in enumerate(centers):
                    distance = 0
                    for i in range(len(point)):
                        distance += (point[i] - center[i]) ** 2
                    distance = distance ** 0.5
                    if distance < min_dis:
                        min_dis = distance 
                        min_index = y
                new_label[x] = min_index
                sum_distance += min_dis
                
            count = [0] * k
            sum_r = [0] * k
            sum_g = [0] * k
            sum_b = [0] * k
            for i in range(len(new_label)):
                label = int(new_label[i])
                count[label] += 1
                sum_r[label] += points[i,0]
                sum_g[label] += points[i,1]
                sum_b[label] += points[i,2]

            for i in range(len(centers)):
                centers[i] = [sum_r[i]/count[i], sum_g[i]/count[i], sum_b[i]/count[i]]

            if new_label == old_label:
                break    
                
            for i in range(len(old_label)):
                old_label[i] = new_label[i]

        if sum_distance < best_distance:
            best_label = new_label
            best_distance = sum_distance


    return best_label
```

```python

k = 5
filename = '/Users/longchen/Downloads/IMG_0023.JPG'
    
im = Image.open(filename)
im = im.resize((100,100), resample=Image.BILINEAR)
im = np.array(im,dtype='float')
points = im.reshape((100*100,3))

labels = kmeans(points, k)

centers = [None] * k
count = [0] * k
sum_r = [0] * k
sum_g = [0] * k
sum_b = [0] * k
for i in range(len(labels)):
    label = int(labels[i])
    count[label] += 1
    sum_r[label] += points[i,0]
    sum_g[label] += points[i,1]
    sum_b[label] += points[i,2]

for i in range(k):
    centers[i] = [sum_r[i]/count[i], sum_g[i]/count[i], sum_b[i]/count[i]]


for x in range(100):
    for y in range(100):
        im[x,y] = centers[int(labels[x * 100 + y])]

plt.imsave("/Users/longchen/Downloads/aaaaa.png", im/255)
```
